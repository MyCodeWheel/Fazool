```python
china_data = df.loc[df.Country == 'China']
```

```python
avgTemp_china = LabelEncoder()
```

```python
cd = china_data.copy()
```

```python
import datetime 
```

```python
data_for_predict = []
year= 2014
month = 1
date = 1
for i in range(10):
    month = 1
    for j in range(12):
        dateee = str(year) + '-' + str(month)+'-'+str(date)
        cadateee = datetime.datetime.strptime(dateee, '%Y-%m-%d')
        data_for_predict.append(cadateee.date())
        month = month+1
    year = year+1
```

```python
cd['AvgTemp'] = avgTemp_china.fit_transform(cd['AverageTemperature'])
```

```python
X = cd['AvgTemp'].values.reshape(-1,1)
Y = cd['AvgTemp'].values.reshape(-1,1)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
decisionTree = DecisionTreeClassifier()
decisionTree.fit(X_train, Y_train)
```

```python
pred = pd.DataFrame(data_for_predict).values.reshape(-1,1)
```

```python
a = decisionTree.predict(pred)
```
